```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import sympy as smp
```

```python
#Using Euler's implicit and explicit functions
#The guess for implcit is taken from explicit function

def Implicit_Explicit(f,y0,t):  
  y = np.zeros(len(t))
  y1=np.zeros(len(t))
      
  y[0] = y0
  y1[0]=y0
  for n in range(0,len(t)-1):
    h=t[n+1]-t[n]
    y[n+1] = y[n] + f(y[n],t[n])*h
    y1[n+1]=y1[n]+ f(y[n+1],t[n+1])*h
    
  return y,y1
```

```python
#Considered the 4 functions with boundary conditions given in the example problems

def main():
  f1 = lambda y,t: (2-2*t*y)/(t**2+1)
  f2 = lambda y,t: (y**2)/(1+t)
  f3 = lambda y,t: (y**2+y)/t
  f4 = lambda y,t: (-t*y)+(4*t/y)
  opt=int(input("Enter the function you wish to choose:\n\n1.(2-2*t*y)/(t**2+1) \n\n2.(y**2)/(1+t)\n\n3.(y**2+y)/t\n\n4.(-t*y)+(4*t/y) "))
  if opt==1:
    h=0.1
    t=np.arange(0,1,h)
    f=f1
    y0=1
    y_exact = ((2*t)+1)/(t**2+1)
  elif opt==2:
    h=0.1
    t=np.arange(1,2,h)
    f=f2
    y0=-1/np.log(2)
    y_exact =-1/np.log(t+1)
  elif opt==3:
    h=0.2
    t=np.arange(1,3,h)
    f=f3
    y0=-2
    y_exact =(2*t)/(1-2*t)
  elif opt==4:
    h=0.1
    t=np.arange(0,1,h)
    f=f4
    y0=1
    y_exact =np.sqrt(4-3*np.exp(-t**2))
  else:
    print("Wrong option selected.\n")
  y_explicit,y_implicit = Implicit_Explicit(f,y0,t)
  plt.figure(figsize=(10,10))
  plt.plot(t,y_implicit,'b.-',t,y_exact,'r-',t,y_explicit,'g.-')
  plt.legend(['Implicit','Exact','Explicit'])
  plt.grid(True)
  plt.show()
```

```python
#Function 2
main()
```

```python
f3 = lambda y,t: (y**2+y)/t
y0=-2
y_exact =(2*t)/(1-2*t)

#For h=0.1
t1=np.arange(1,3.1,0.1)
y_exact1 =(2*t1)/(1-2*t1)

#For h=0.2
t2=np.arange(1,3.1,0.2)
y_exact2 =(2*t2)/(1-2*t2)

#For h=0.4
t3=np.arange(1,3.1,0.4)
y_exact3 =(2*t3)/(1-2*t3)

#For h=0.5
t4=np.arange(1,3.1,0.5)
y_exact4 =(2*t4)/(1-2*t4)

y_explicit1,y_implicit1 = Implicit_Explicit(f3,y0,t1)
y_explicit2,y_implicit2 = Implicit_Explicit(f3,y0,t2)
y_explicit3,y_implicit3 = Implicit_Explicit(f3,y0,t3)
y_explicit4,y_implicit4 = Implicit_Explicit(f3,y0,t4)

y_ie1=y_implicit1-y_exact1
y_ie2=y_implicit2-y_exact2
y_ie3=y_implicit3-y_exact3
y_ie4=y_implicit4-y_exact4

y_ee1=y_explicit1-y_exact1
y_ee2=y_explicit2-y_exact2
y_ee3=y_explicit3-y_exact3
y_ee4=y_explicit4-y_exact4

#Plotting implicit errors
plt.figure(figsize=(10,10))
plt.plot(t1,y_ie1,'b.-',t2,y_ie2,'r.-',t3,y_ie3,'g.-',t4,y_ie4,'y.-')
plt.legend(['h=0.1','h=0.2','h=0.4','h=0.5'])
plt.grid(True)
plt.title("Implicit errors")

#Plotting explicit errors
plt.figure(figsize=(10,10))
plt.plot(t1,y_ee1,'b.-',t2,y_ee2,'r.-',t3,y_ee3,'g.-',t4,y_ee4,'y.-')
plt.legend(['h=0.1','h=0.2','h=0.4','h=0.5'])
plt.grid(True)
plt.title("Explicit errors")
plt.show()
plt.show()
```
